```python
from qiskit.visualization import plot_state_qsphere
from qiskit.visualization import plot_bloch_multivector
```

```python
#Create a simple circuit with just one qubit
qc = QuantumCircuit(1)
#Get the state vector simulator backend
statevector_simulator = Aer.get_backend('statevector_simulator')
#Run the circuit and get the state vector of the qubit
result = execute(qc, statevector_simulator).result()
statevector_results = result.get_statevector(qc)
```

```python
plot_bloch_multivector(statevector_results)
```

```python
plot_state_qsphere(statevector_results)
```

```python
qc = QuantumCircuit(1)
qc.x(0)
#Execute the circuit
result = execute(qc, statevector_simulator).result()
statevector_results = result.get_statevector(qc)
plot_state_qsphere(statevector_results)
```

```python
qc = QuantumCircuit(1)
qc.h(0)
#Execute the circuit
result = execute(qc, statevector_simulator).result()
statevector_results = result.get_statevector(qc)
plot_bloch_multivector(statevector_results)
```

```python
plot_state_qsphere(statevector_results)
```
